```python
filepath = '/content/drive/MyDrive/AIM_paper/AIM_Paper_DATA_Analysis/movie/K-001.avi'
video = cv2.VideoCapture(filepath) #'' 사이에 사용할 비디오 파일의 경로 및 이름을 넣어주도록 함

if not video.isOpened():
    print("Could not Open :", filepath)
    exit(0)
```

```python
#불러온 비디오 파일의 정보 출력
length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)

print("length :", length)
print("width :", width)
print("height :", height)
print("fps :", fps)
```

```python
while(video.isOpened()):
    ret, image = video.read()
    if(int(video.get(1)) % fps == 0): #앞서 불러온 fps 값을 사용하여 1초마다 추출
        cv2.imwrite(filepath[:-4] + "/frame%d.jpg" % count, image)
        print('Saved frame number :', str(int(video.get(1))))
        count += 1
        
video.release()
```

```python
images = glob.glob('/content/drive/MyDrive/AIM_paper/AIM_Paper_DATA_Analysis/images/*.jpg')
```

```python
images
```

```python
histogram_dict = {}
for i in images:
  image = cv2.imread(i, cv2.IMREAD_COLOR) # openCV 로 칼러이미지 
  image_cut = image[420:840, 776:1237]
  dst = cv2.cvtColor(image_cut, cv2.COLOR_BGR2GRAY)
  print(image_cut.shape)
  cv2_imshow(image_cut)
  hist,bins = np.histogram(dst.flatten(),256,[0,256])
  plt.hist(dst.flatten(),256,[0,256], color = 'b')
  plt.xlim([0,256])
  plt.show()
  contrast = 1000 * dst.var()
  print(contrast)
  print(i)
  histogram_dict[i] = (contrast)
```

```python
print(histogram_dict)
```

```python
for k, v in histogram_dict.items():
  print("images: {}, hysteresis: {}". format(k, v))
```

```python
max(histogram_dict,key=histogram_dict.get)
[k for k,v in histogram_dict.items() if max(histogram_dict.values()) == v]
```
